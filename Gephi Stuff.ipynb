```python
import pywikibot as pwb
import pandas as pd
import networkx as nx
```

```python
langs = ['en','zh','tr','ja','ru']
```

```python
def translate(site,title):
    page = pwb.Page(site, title)

    for link in page.iterlanglinks():
        if link.site.code == 'en':
            return link.title
        
    return title
```

```python
for lang in langs:
    print '*'*20
    print lang
    print '*'*20
    df = pd.read_csv('%s25 [Nodes].csv' % lang)
    site = pwb.Site(lang, 'wikipedia')
    for i,row in df.sort('Eigenvector Centrality', ascending=False).head(100).iterrows():
        if lang is not 'en':
            print translate(site, row['Label'].decode('utf-8')), row['Eigenvector Centrality']
        else:
            print row['Label'], row['Eigenvector Centrality']
```
